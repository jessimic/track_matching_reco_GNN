```python
import sys
import yaml
sys.path.insert(0, '/sdf/data/neutrino/software/spine/src/')
from spine.driver import Driver
spine_file="MiniProdN5p1_NDComplex_FHC.spine.full.sanddrift.0002471.MLRECO_SPINE.hdf5"
cfg = f'''
        base:
          verbosity: warning
        build:
          mode: both
          fragments: false
          particles: true
          interactions: true
          units: cm
        io:
          reader:
            file_keys: {spine_file}
            skip_unknown_attrs: True
            name: hdf5
        '''
driver  = Driver(yaml.safe_load(cfg))
nd_spill_idx=0
interaction_dict = {}
spine_data = driver.process(nd_spill_idx)
flagged_key = []
for ixn in spine_data['truth_interactions']:
    idkey = str(ixn.id)
    t_adj = round(float(ixn.t), 3)
    interaction_dict[idkey] = ixn.interaction_id
    if t_adj in [1235.837]: #, 1349.141]:
        print(f"vtxTime={t_adj} track_id={ixn.track_id}, interaction={ixn.interaction_id}")
        flagged_key.append(idkey)
        
        
for p in spine_data['truth_particles']:
    if str(p.interaction_id) not in flagged_key:
        continue
    #particle.interaction_id tells you int from 0 that corresponds to the vertex ID list in interaction.interaction_id
    print("Interaction:", interaction_dict[str(p.interaction_id)]) 
    print("PDG:", p.pdg_code)
    print("Track ID:", p.track_id)
```

```python
import uproot
import numpy as np

tms = uproot.open("MiniProdN5p1_NDComplex_FHC.tmsreco.full.sanddrift.0002471.TMSRECO.root")

spill = 0
vtx_idx = 69  # The selected vertex
target_time = 1235.837

print("=== Analysis for selected vertex ===")
print(f"Vertex index: {vtx_idx}, Target time: {target_time}\n")

# Step 1: Get RecoTrackPrimaryParticleIndex from TruthInfo
reco_track_primary_idx = tms["Truth_Info"]['RecoTrackPrimaryParticleIndex'].array(library="np")[spill]
print(f"1. TruthInfo['RecoTrackPrimaryParticleIndex'] at vtx_idx {vtx_idx}: {reco_track_primary_idx[vtx_idx]}")

# Step 2: Switch to Truth_Spill tree and get Parent info
parents = tms["Truth_Spill"]['Parent'].array(library="np")[spill]
print(f"\n2. Truth_Spill['Parent'] at vtx_idx {vtx_idx}: {parents[vtx_idx]}")

# Step 3: Find candidate secondary particles in the tree
track_ids = tms["Truth_Spill"]['TrackId'].array(library="np")[spill]
pdg_codes = tms["Truth_Spill"]['PDG'].array(library="np")[spill]
birth_pos_x = tms["Truth_Spill"]['BirthPosition'].array(library="np")[spill][:, 0]
birth_pos_y = tms["Truth_Spill"]['BirthPosition'].array(library="np")[spill][:, 1]
birth_pos_z = tms["Truth_Spill"]['BirthPosition'].array(library="np")[spill][:, 2]
death_pos_x = tms["Truth_Spill"]['DeathPosition'].array(library="np")[spill][:, 0]
death_pos_y = tms["Truth_Spill"]['DeathPosition'].array(library="np")[spill][:, 1]
death_pos_z = tms["Truth_Spill"]['DeathPosition'].array(library="np")[spill][:, 2]

parent_track_id = parents[vtx_idx]
secondary_particles = np.where(track_ids == parent_track_id)[0]

print(f"\n3. Secondary particles with Parent TrackId={parent_track_id}:")
for idx in secondary_particles:
    print(f"\n   Index: {idx}")
    print(f"   TrackId: {track_ids[idx]}")
    print(f"   PDG: {pdg_codes[idx]}")
    print(f"   BirthPosition: ({birth_pos_x[idx]:.3f}, {birth_pos_y[idx]:.3f}, {birth_pos_z[idx]:.3f})")
    print(f"   DeathPosition: ({death_pos_x[idx]:.3f}, {death_pos_y[idx]:.3f}, {death_pos_z[idx]:.3f})")
```
